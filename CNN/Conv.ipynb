```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
 inChans = 3
outChans = 15
krnSize = 5
stride = 1
padding = 0

c = nn.Conv2d(inChans,outChans,krnSize,stride,padding)

print(c)
print(c.weight.shape)
print(c.bias.shape)
```

```python
imgSize = (1,3,64,64)

img = torch.rand(imgSize)

img2view = np.squeeze(img.permute(2,3,1,0).numpy())
plt.imshow(img2view)
plt.show()
```

```python
convImgs = c(img)

print(img.shape)
print(convImgs.shape)
```

```python
fig,axs = plt.subplots(3,5,figsize=(10,5))

for i,ax in enumerate(axs.flatten()):
    I = torch.squeeze(convImgs[0,i,:,:]).detach()
    ax.imshow(I)

plt.tight_layout()
plt.show()
    
```
