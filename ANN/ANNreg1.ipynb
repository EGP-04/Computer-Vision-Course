```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
N = 30
x = torch.randn(N,1)
y = x + torch.randn(N,1)

plt.plot(x,y,'s')
plt.show()
```

```python
ANNreg = nn.Sequential(
    nn.Linear(1,1),
    nn.ReLU(),
    nn.Linear(1,1)
)
ANNreg
```

```python
learningRate = 0.05

lossFunc = nn.MSELoss()

optimizer = torch.optim.SGD(ANNreg.parameters(),lr = learningRate)
```

```python
numEpoch = 500
losses = torch.zeros(numEpoch)

for epochi in range(numEpoch):
    yHat = ANNreg(x)

    loss = lossFunc(yHat,y)
    losses[epochi] = loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
 predictions = ANNreg(x)

testLoss = (predictions-y).pow(2).mean()

plt.plot(losses.detach())
plt.show()
```
