```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(x):
  return 3*x**2 - 3*x + 4
    
def deriv(x):
  return 6*x - 3
```

```python
x = np.linspace(-2,2,2001)

plt.plot(x,fx(x), x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
     
```
